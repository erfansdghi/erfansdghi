```python
class dnode:

  def __init__(self,data):
    self.data = data
    self.next = None
    self.prev = None

class dlinkedlist:

  def __init__(self):
    self.head = None

  def display(self):
    t = self.head
    while (t != None):
      print(t.data, end='<-->')
      t = t.next
    print('NONE')

  def addinstart(self,data):
    n = dnode(data)
    if (self.head == None):
      self.head = n
    else:
      n.next = self.head
      self.head.prev = n
      self.head = n

  def addinend(self,data):
    n = dnode(data)
    if (self.head != None):
      t = self.head
      while (t.next != None):
        t = t.next
      t.next = n
      n.prev = t
    else:
      self.head = n

  def addafter(self,m,data):
    n = dnode(data)
    t = self.head
    while (t.data != m):
      t = t.next
    n.next = t.next
    n.prev = t
    t.next = n
    t.next.prev = n

  def delfirst(self):
    if (self.head == None):
      return -1
    self.head = self.head.next
    self.head.prev = None

  def dellast(self):
    if (self.head == None):
      return -1
    t = self.head
    while(t.next != None):
      t = t.next
    t.prev.next = None
    t.prev = None

  def delete(self,m):
    t = self.head
    while (t.data != m):
      t = t.next
    t.next.prev = t.prev
    t.prev.next = t.next
    t.next = None
    t.prev = None

  def delmid(self):
    t = self.head
    count = 0
    while (t.next != None):
      t = t.next
      count += 1
    t = self.head
    if (count % 2 == 0):
      for i in range(count//2):
        t = t.next
      t.next.prev = t.prev
      t.prev.next = t.next
      t.next = None
      t.prev = None
    else:
      for i in range((count//2) + 1):
        t = t.next
      t.next.prev = t.prev
      t.prev.next = t.next
      t.next = None
      t.prev = None


# Create a doubly linked list
my_doubly_linked_list = dlinkedlist()

# Add nodes to the start
my_doubly_linked_list.addinstart(3)
my_doubly_linked_list.addinstart(2)
my_doubly_linked_list.addinstart(1)

# Display the doubly linked list
my_doubly_linked_list.display()

# Add nodes to the end
my_doubly_linked_list.addinend(4)
my_doubly_linked_list.addinend(5)

# Display the doubly linked list
my_doubly_linked_list.display()

# Add a node after a specific value
my_doubly_linked_list.addafter(3, 6)

# Display the doubly linked list
my_doubly_linked_list.display()

# Delete the first node
my_doubly_linked_list.delfirst()

# Display the doubly linked list
my_doubly_linked_list.display()

# Delete the last node
my_doubly_linked_list.dellast()

# Display the doubly linked list
my_doubly_linked_list.display()

# Delete a specific value
my_doubly_linked_list.delete(3)

# Display the doubly linked list
my_doubly_linked_list.display()

# Delete the middle node
my_doubly_linked_list.delmid()

# Display the doubly linked list
my_doubly_linked_list.display()
```
